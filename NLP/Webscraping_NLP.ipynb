## Task: 
### Fetch data from php.net website, clean it and give frequency of each word.

```python
#Importing required python libraries
from bs4 import BeautifulSoup
import urllib.request
import nltk
import re
import matplotlib.pyplot as plt
from wordcloud import WordCloud
```

```python
res=urllib.request.urlopen('http://php.net/')
```

```python
html=res.read()
soup=BeautifulSoup(html,'html.parser')
```

```python
text=soup.get_text()
```

```python
#Replacing all new line char \n with space
text=text.replace('\n',' ').strip()
```

```python
#Replacing multiple spaces with single space
text=re.sub(r'\s+',' ',text)
```

```python
#Extracting only characters from the text and removing digits and other punctuations.
text=re.sub(r'[^a-zA-Z ]','',text)
```

```python
#Splitting the sentences to list of words and converting all words to lowercase
text_list=[x.lower() for x in text.split()]
```

```python
stop_words=nltk.corpus.stopwords.words('english')
```

```python
#Removing all words with length less than or equal to 2 or are stopwords
words=[]
for word in text_list:
    if(len(word)<=2 or word in stop_words):
        continue
    else:
        words.append(word)
```

```python
extracted_words=' '.join(words)
```

```python
wordcloud=WordCloud(width = 1000, height = 600,collocations=False).generate(extracted_words)
```

```python
# plot the WordCloud image                        
plt.figure(figsize = (8, 8), facecolor = None) 
plt.imshow(wordcloud) 
plt.axis("off") 
plt.tight_layout(pad = 0) 
  
plt.show()
```

```python
freq = nltk.FreqDist(words)
```
